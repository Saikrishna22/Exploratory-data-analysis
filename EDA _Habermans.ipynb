```python
#importing all the required libraires 
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('Haberman.csv',names=(['age','years','posnodes','survstatus']))
```

```python
df.describe()
```

```python
df.info()
```

```python
df.survstatus.value_counts()
```

```python
df['survstatus']=df['survstatus'].map({1:'survived',2:'died'})
```

```python
sns.distplot(df['age'],bins=30)
```

```python
sns.jointplot(x='posnodes',y='age',kind='scatter',data=df)
```

```python
sns.pairplot(df)
```

```python
sns.jointplot(x='age',y='years',data=df,kind='scatter')
```

```python
sns.jointplot(x='age',y='years',data=df,kind='reg')
plt.title("years vs age")
```

```python
sns.FacetGrid(df, hue = 'survstatus', size = 6).map(plt.scatter, 'posnodes', 'years').add_legend()
plt.show()
```

```python
sns.FacetGrid(df, hue = 'survstatus', size = 6).map(plt.scatter, 'age', 'years').add_legend()
plt.show()
```

```python
sns.FacetGrid(df, hue='survstatus', size = 5).map(sns.distplot, 'posnodes').add_legend();
plt.show();
```

```python
#here we can conclude that - Patients having 0 axil nodes are more likely to survive
```

```python
sns.FacetGrid(df, hue='survstatus', size = 5).map(sns.distplot, 'age').add_legend()
plt.show();
```

```python
#here in the above histogram we can say that age below 40 are more likely to survive.
```

```python
sns.boxplot(x='survstatus',y='posnodes',data=df)
```

```python
###this shows that axil nodes with less that 1 have survived
#and greater than 1 has less chance of surviving.
```

```python
sns.violinplot(x='survstatus', y='posnodes', data = df, size = 9)
```

```python
counts,bin_edges = np.histogram(df['posnodes'],bins = 30, density = True)
pdf = counts/(sum(counts))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)



plt.xlabel('axil nodes')
plt.grid()

plt.show()
```

```python
#pdf and cdf for axil nodes
```

```python
#from all the above plots we can get to know that Axil nodes(posnodes) is the main feature to find the survival status of a patient 
```

```python
sns.violinplot(x='survstatus', y='posnodes', data = df, size = 9)
```

```python
#from this violin plot we can conirm that most of the people with axil nodes(posnodes) > 0 are more likely to die..
```

```python
#Final conclusion
#1. the majority of operations are performed on people age range between 38 and 68.
#2.  large number of operation were done in span of 7 years between 1960 and 1966 .
#3. patients with 0 axil nodes are more likely to survive irrespective to there age.
#4. It is very much less likely to have patients with axil nodes more than 30.

#Feature importance:
#1. Axil nodes is most important feature in this dataset.
#2. Age is also the important feature in the dataset as patients who aged less than 40 are more likely to survive inspite of having axil node >= 1
```
