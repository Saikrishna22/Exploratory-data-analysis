```python
%matplotlib inline
%matplotlib notebook
```

```python
#Creating an dataframe from Iris csv file
df=pd.read_csv('Iris.csv')
```

```python
#let us take the Species column as the responce value(dependent variable) and the other four columns as feature variable(independent variable)
```

```python
#Our objective here is to predict a particular species with the given independent variables,so lets analys the dataset using seaborn
```

```python
df.head()
```

```python
#we dont need the id column here were we already have index in our data frame
```

```python
df1=df.drop('Id',axis=1)
```

```python
df1.head()
```

```python
df.head()
```

```python
df1.plot(x='SepalLengthCm',y='SepalWidthCm',kind='scatter')
```

```python
df1.plot(x='PetalLengthCm',y='PetalWidthCm',kind='scatter')
```

```python
#from the above figures we can know that , prediction can be done for the size of th epetal and sepal and not for the particular species.
# so we use Seaborn to differentiate the species of Iris.
```

```python
sns.FacetGrid(df1, hue='Species',size=5).map(plt.scatter,'SepalLengthCm', 'SepalWidthCm').add_legend()
# this is for Sepal Length and Sepal Width
```

```python
sns.FacetGrid(df1, hue='Species',size=5).map(plt.scatter,'PetalLengthCm', 'PetalWidthCm').add_legend()
#this is for petal lenght and petal width
```

```python
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=df1, size=5, kind='reg',color='green')
```

```python
#we can use kind = hex too for better visualization
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=df1, size=5, kind='hex',color='purple')
```

```python
# pairplot shows the bivariate relation between each pair of features.
#  this pairplot gives us different scenarios that we can predict the required species 
```

```python
sns.pairplot(data=df1, hue="Species", size=3,palette='spring',markers=["o", "s", "D"])
#this is a bivariate analysis
```

```python
# Lets now see univariate analysis 
#box plots and violin plots
```

```python
# the violin plot consist of both boxplot and pdf plot in it.
plt.figure()
sns.violinplot(x="Species", y="SepalWidthCm", data=df1,palette='rainbow')
plt.show
```

```python
#box plot shows us the 25% line, median and 75% line clearly, this is helpful in catagorizing different species with accurate percentile.

plt.figure()
sns.boxplot(x="Species", y="SepalWidthCm", data=df1,palette='rainbow')
plt.show()
```

```python
plt.figure()
sns.boxplot(y="SepalLengthCm", x="Species",data=df1, palette="coolwarm")
plt.show()
```

```python
#simple regression plot
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=df1, size=6, kind='reg',color='purple')
```
